### 3D Contour Plots

```python
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2)) # sine wave of circle.

x = np.linspace(-6, 6, 30)  # linearly spaced line, -6 to +6, 30 points.
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
# Plotting - MatLab style

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
# ax.set_xlabel('x')
# ax.set_ylabel('y')
# ax.set_zlabel('z')

# Increase viewing tilt and rotate 
# ax.view_init(60, 35);
```

# MATLAB style subplots - plt


```python
# MATLAB style - plt interface

fig4 = plt.figure(figsize=(10, 5)) # create a plot figure
# create the first of two panels and set current axis
ax1 = plt.subplot(1, 3, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
ax2 = plt.subplot(1, 3, 2)
plt.plot(x, np.cos(x));

# Add 3rd subplot
ax3 = plt.subplot(1, 3, 3)
plt.plot(x, 2*np.sin(x))

print(fig4, ax1, ax2, ax3)
```
